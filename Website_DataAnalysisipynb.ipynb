```python
WINDOW_SIZE = 12
df['rolling_mean'] = df.Close.rolling(window=WINDOW_SIZE).mean()
df['rolling_std'] = df.Close.rolling(window=WINDOW_SIZE).std()
```

```python
WINDOW_SIZE = 12
df['vrolling_mean'] = df.Volume.rolling(window=WINDOW_SIZE).mean()
df['vrolling_std'] = df.Volume.rolling(window=WINDOW_SIZE).std()
```
